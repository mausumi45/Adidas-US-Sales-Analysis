### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
warnings.filterwarnings('ignore')
```

### Importing Datasets

```python
data = pd.read_excel("Adidas US Sales Datasets.xlsx", skiprows = 4)
data.drop("Unnamed: 0", axis = 1,inplace = True )
data
```

### Shape ,size  & Dimension of dataset

```python
data.shape
```

```python
data.size
```

```python
data.ndim
```

### Renaming the column

```python
data = data.rename(columns = {'Units Sold' : 'Quantity', 'Operating Profit' : 'Profit', 'Operating Margin' : 'Margin'})
data
```

### Datatype of columns

```python
data.dtypes
```

### Extracted month name and year column from invoice date

```python
data['month'] = data['Invoice Date'].dt.month_name()
data['year'] = data['Invoice Date'].dt.year
```

```python
data
```

### Missing Values 

```python
data.isnull().sum()
```

### Data Analysis

### Correlation Analysis

```python
sns.heatmap(data.corr(), cmap="crest", annot=True)
```

####  Quantity, Total sales and profits are highly correlated to each other.


### Regionwise top selling & profitable Product

```python
Region_top_prod = data.groupby(['Region','Product'])['Total Sales','Profit'].sum()
Region_top_prod['sales_rank'] = Region_top_prod.groupby('Region')['Total Sales'].rank(ascending = False)
Region_top_prod['profit_rank'] = Region_top_prod.groupby('Region')['Profit'].rank(ascending = False)
Region_top_prod = Region_top_prod.loc[(Region_top_prod['sales_rank'] == 1.0) & (Region_top_prod['profit_rank'] == 1.0)].reset_index()
Region_top_prod
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Region', y='Total Sales', hue='Product', data = data, palette='Set2')
plt.title('Top Selling and Profitable Products by Region', fontsize = 20)
plt.xlabel('Region')
plt.ylabel('Sales (in thousands)')
plt.show()
```

#### Men's Street Footware is the most profitable product in every region of US .
#### The highest sales of men's Footware occured in southern region.

### Top 5 Profitable state

```python
Top_5_state = data.groupby('State')['Profit'].sum().sort_values(ascending = False).reset_index()
Top_5_state = Top_5_state[['State', 'Profit']].head(5)
Top_5_state
```

```python
sns.set(rc={'figure.figsize':(8,6)})
sns.barplot(x='Profit', y='State', data=Top_5_state, palette='YlOrBr')
plt.title('Top 5 Profitable States', fontsize = 20)
plt.xlabel('Profit (in thousands)')
plt.ylabel('State')
plt.show()
```

#### Newyork was the most profitable city.

### Bottom 5 Profitable state

```python
Bottom_5_state = data.groupby('State')['Profit'].sum().sort_values(ascending = True).reset_index()
Bottom_5_state = Bottom_5_state[['State', 'Profit']].head(5)
Bottom_5_state
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x='Profit', y='State', data=Bottom_5_state, palette='cubehelix')
plt.title('Bottom 5 Profitable States', fontsize = 20)
plt.xlabel('Profit (in thousands)')
plt.ylabel('State')
plt.show()
```

#### - Nebraska was the lowest profitable state.

### Top 5 Profitable city

```python
Top_5_City = data.groupby('City')['Profit'].sum().sort_values(ascending = False).reset_index()
Top_5_City = Top_5_City[['City', 'Profit']].head(5)
Top_5_City
```

```python
sns.set(rc={'figure.figsize':(8,6)})
sns.barplot(data=Top_5_City, x="City", y="Profit", palette = 'mako')
plt.title('Top 5 Profitable States', fontsize = 20)
plt.xlabel('Profit')
plt.ylabel('City')
plt.show()
```

#### -Most Profitable city was Charleston.

### Monthwise profit  & quantity anlysis


```python
Monthwise_profit = data.groupby('month')['Profit', 'Quantity'].sum()
Monthwise_profit
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(x='month', y='Profit', data=Monthwise_profit, color='blue', label = 'Profit')
plt.title('Month-wise Profit Analysis', fontsize = 20)
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

#### - Most Profitable month was august  and least profitable month was march.

### Yearwise profit analysis product wise

```python
yearwise_product_profit = data.groupby(['year', 'Product'])['Profit'].sum().reset_index()
yearwise_product_profit
```

```python
df_pivot = yearwise_product_profit.pivot_table(index='Product', columns='year', values='Profit')
sns.set(rc={'figure.figsize':(20,10)})
df_pivot.plot(kind='bar', stacked=True)

plt.title('Product-wise Profit Analysis in Each Year', fontsize = 20)
plt.xlabel('Product')
plt.ylabel('Profit')
plt.legend(title='Year')
plt.show()
```

#### - Men's Street Footware was the most selling product in 2021 and 2020.
#### - Women's Athletic Footware was the least selling product .

### Top profitable product in each sales method

```python
methodwise_profitable_prod = data.groupby(['Sales Method','Product'])['Profit'].sum().reset_index()
methodwise_profitable_prod['rank'] = methodwise_profitable_prod.groupby('Sales Method')['Profit'].rank(ascending = False)
methodwise_profitable_prod.loc[methodwise_profitable_prod['rank'] == 1.0]
```

### Productwise margin analysis

```python
prod_margin = data.groupby('Product')['Margin'].mean().reset_index()
prod_margin = prod_margin.sort_values(by = 'Margin',ascending = False)
prod_margin
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.set_palette('Spectral')
plt.pie(prod_margin['Margin'], labels=prod_margin['Product'], autopct='%1.1f%%')
plt.title('Product-wise Margin Analysis', fontsize = 15)
plt.show()
```

### Yearwise retailer sales analysis

```python
retailer_sales = data.groupby(['Retailer','year'])['Total Sales'].mean().reset_index()
retailer_sales
```

```python
retailer_sales = retailer_sales.set_index('Retailer')
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x=retailer_sales.index, y='Total Sales', hue='year', data=retailer_sales, palette='rocket')
plt.title('Year-wise Retailer Sales Analysis')
plt.xlabel('Retailer')
plt.ylabel('Sales')
plt.legend(title='Year')
plt.show()
```

### Quarterly sales analysis by each retailer

```python
data['quarter'] = pd.PeriodIndex(data['Invoice Date'], freq = 'Q')
data
```

```python
quarterly_sales_retailer = data.groupby(['Retailer','quarter'])['Total Sales'].sum().reset_index()
quarterly_sales_retailer['rank'] = quarterly_sales_retailer.groupby('Retailer')['Total Sales'].rank(ascending = False)
quarterly_sales_retailer = quarterly_sales_retailer.loc[quarterly_sales_retailer['rank'] == 1.0]
quarterly_sales_retailer
```

```python
sns.set(rc={'figure.figsize':(20,15)})
sns.color_palette("viridis", as_cmap=True)
sns.barplot(x='Retailer', y='Total Sales', hue='quarter', data=data)
plt.title('Quarterly Sales by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Sales (in thousands)')
plt.show()
```
